## Notebook used to merge seasonalities of ingredients. 

```python
import pandas as pd
import numpy as np
import random
import time
import math
```

```python
df_saison = pd.read_csv('data/saison.csv')
df = pd.read_csv('data/data_mock_ingrédients.csv')
```

```python
df_saison.head()
```

```python
df.head()
```

```python
df_saison.columns = ['Nom', 'Type', 'Mois', 'Saison']
```

```python
df_saison = df_saison.drop(columns = ['Mois'])
```

```python
df_saison = df_saison.drop_duplicates()
```

```python
df_saison
```

```python
df_saison['ingredient_reco'] = df_saison['Nom'].str.lower()
```

```python
df_bis = pd.merge(df,df_saison, on='ingredient_reco', how='left')
```

```python
df_bis['saison'] = df_bis['Saison']
```

```python
df_bis
```

```python
df_bis = df_bis.drop(columns = ['Nom', 'Type', 'Saison'])
```

```python
df_bis.to_csv('data/merged_mock_ingrédient.csv')
```
